# Dataset Exploration

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
import seaborn as sns
```

```python
dataset_TAD66K_path = '/home/zerui/SSIRA/dataset/TAD66K/'

# check the dataset 
if not os.path.exists(dataset_TAD66K_path):
    print('Dataset not found.)')

# how many images in the dataset
print('Total number of images in the dataset:', len(os.listdir(dataset_TAD66K_path)))

# labels
label_merge_TAD66K_test = pd.read_csv('/home/zerui/SSIRA/dataset/TAD66K/labels/merge/test.csv')
label_merge_TAD66K_train = pd.read_csv('/home/zerui/SSIRA/dataset/TAD66K/labels/merge/train.csv')

label_unmerge_TAD66K_test_path = '/home/zerui/SSIRA/dataset/TAD66K/labels/unmerge/test/'
label_unmerge_TAD66K_train_path = '/home/zerui/SSIRA/dataset/TAD66K/labels/unmerge/train/'

# there are different categories of labels for the unmerged label, each is a csv file


# check the merged label
print('Number of images in the merged label (train):', len(label_merge_TAD66K_train))
print('Number of images in the merged label (test):', len(label_merge_TAD66K_test))

# check the unmerged label
print('Number of categories of labels in the unmerged label (train):', len(os.listdir(label_unmerge_TAD66K_train_path)))
print('Number of categories of labels in the unmerged label (test):', len(os.listdir(label_unmerge_TAD66K_test_path)))

# check the number of images in each category of the unmerged label
print('Number of images in each category of the unmerged label (train):')
for category in os.listdir(label_unmerge_TAD66K_train_path):
    print(category, len(pd.read_csv(label_unmerge_TAD66K_train_path + category)))

print('Number of images in each category of the unmerged label (test):')
for category in os.listdir(label_unmerge_TAD66K_test_path):
    print(category, len(pd.read_csv(label_unmerge_TAD66K_test_path + category)))

```

```python
import random
import matplotlib.image as mpimg

fig = plt.figure(figsize=(20,4))

for i in range(1,6):
    fig.add_subplot(1, 5, i)
    random_img = random.choice(os.listdir(dataset_TAD66K_path))
    image = mpimg.imread(os.path.join(dataset_TAD66K_path, random_img))
    plt.imshow(image)
    plt.title(random_img)
    plt.axis('off')
plt.show()
```
